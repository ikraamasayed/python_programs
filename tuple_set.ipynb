##### list VS Tuple Comparison


```python
import sys
my_list = [0,1,2,"hello",True]
my_tuple = (0,1,2,"hello",True)
print("LIST SIZE " ,sys.getsizeof(my_list),"bytes")
print("TUPLE SIZE",sys.getsizeof(my_tuple),"bytes")

```

```python
import timeit
print("LIST: ",timeit.timeit(stmt="[0,1,2,3,4,5]",number=1000000))  #create list 1 million times with stmt
print("TUPLE:",timeit.timeit(stmt="(0,1,2,3,4,5)",number=1000000))  #create tuple 1 million times with stmt
```

### Set
all are the unique elements ,
if not unique ignores the repeatation & fetch randomly not in sequence & indexing is not supported
do not support mutable elements but sets are mutable

```python
s= {10,15,61,3,45,21,4,12,41}
s
```

```python
wanna_be_empty_set ={}# if try to create empty set like this
print(type(wanna_be_empty_set)) # this will show empty dictionary

actual_empty_set=set() # defining the set using " set()" just like " type() "
print(type(actual_empty_set)) # this will return the empty set
```

```python
s.add(1)

for i in range(9):
    s.add(i)
print(s)
```

#### DISCARD / remove but if not found will not raise any error

```python
print(s)
for el in list(s ):
  if el not in range(1,9):
    s.discard(el)
print(s)
```

#### .remove(el) : remove an element "el"

```python
s.remove(0)
s.discard(0)
print(s,len(s))

```

#### .pop() : removes the random Value from the set

```python
s.pop()
print(s,len(s))
```

#### .clear() : empties the whole element

```python
s.clear()
print("set s :", s)
```

```python
a= {1,3,5,7,9,11,13,17,19}
print()
b={1,2,3,4,5,6,7,8,9}
print("set a : ",a,"    ","set b : ",b)
print("intersection of a & b : ",a.intersection(b))
print("union of a & b : ",a.union(b))

```

#### Difference between (symmetric_difference())

```python
diff = a.difference(b)  # return the different/compared value from , here a has 11,13,17,19 different
print("difference of a & b : ",diff)
ffid = b.difference(a)  # here b has 2,4,6,8
print("difference of a & b : ",ffid)
```

```python
sym_diff = b.symmetric_difference(a)
print("symmetric difference of a & b : ",sym_diff)
mys_diff = a.symmetric_difference(b)
print("symmetric difference of a & b : ",mys_diff)
```

#### TYPES OF UPDATES

```python
a.difference_update(b)
print(a)
```

```python
a.intersection_update(b)
print(a)
```

```python
a.symmetric_difference_update(b)
print(a)
```
