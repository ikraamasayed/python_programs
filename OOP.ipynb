**Q . Create Account class with 2 attributes - balance & account no. Create methods for debit, credit & printing the balance**

````python
class Account:
    
    def __init__(self,balance,acc_no):
        self.balance=balance
        self.acc_no=acc_no

    def debit(self,amount):
        self.balance -= amount
        print("$",amount," was debitted total balance is ",self.get_balance())
    
    def credit(self,amount):
        self.balance+=amount
        print("$",amount,"was creditted total balance is ",self.balance)

    def get_balance(self):
        return self.balance

acc_1= Account(1500,247)
acc_1.credit(1000)
acc_1.debit(500)
        
````

✅ Pros: use get_balace() instead of self.balance 

Encapsulation: If get_balance() is modified (e.g., balance is retrieved from a database or calculated dynamically), you don’t need to modify the methods that call it.
Better abstraction: It hides the direct attribute access, making it easier to change the internal logic without affecting other parts of the class.

❌ Cons:

Slightly slower due to an extra function call (but this is negligible unless performance is critical)

##### Super Method
* How to use `super()` in Inheritance
* can use to call class as well as method to sub class 
* child(sub)class search for sub class's `__init__` if its not there then it'll go for the parent(super)class `__init__`
* super() method is used to access `__init__` methods of the parent(super) class in presence of child(sub) class's `__init`
* Method Resolution Order (MRO) always goes from Left TO Right 

```class Car:
    def __init__(self,type):→ → → → →→⬎                   
        self.type = type               ↓ 
                                       ↓
    @staticmethod                      ↓
    def start():                       ↓
        return "Car Start's...."       ↓
                                       ↓
    @staticmethod                      ↓
    def Off():                         ↓
        return "Car Turned Off..."     ↓
                                       ↓ 
class Toyota (Car):                    ↓
    def __init__(self, name,type):     ↓
        super().__init__(type) ← ← ← ← ↲ 
        self.name = name

    #instead of using argument "type" in child class which is also availble at parent class we use super() to call the parent class argument 

gadi1 = Toyota("Fortuner", "patrol")
print(gadi1.type)```

````python
class Car:
    def __init__(self,type):
        self.type = type

    @staticmethod
    def start():
        return "Car Start's...."
    
    @staticmethod
    def Off():
        return "Car Turned Off..."
    
class Toyota (Car):
    def __init__(self, name,type):
        super().__init__(type)
        super().start
        self.name = name

gadi1 = Toyota("Fortuner", "patrol")
print(gadi1.type)
print(gadi1.start())
````

### Polymorphism : Operator Overloading
When the same operator is allowed to have different meaning according to the context(no,str,list).


````python
class Complex:
    def __init__(self,real,img):
        self.real= real
        self.img= img

    def showNumber(self):
        print(self.real,"i +",self.img,"j")

    def __add__(self,num):
        NEWREAL =   self.real + num.real
        NEWIMG = self.img + num.img
        return Complex(NEWREAL,NEWIMG)
    
    def __sub__(self,num):
        NEWREAL = self.real - num.real
        NEWIMG = self.img - num.img
        return Complex(NEWREAL , NEWIMG)
        

num1= Complex(2,5)
num1.showNumber()

num2=Complex(4,7)
num2.showNumber()

num3 = Complex(6,9)
num3.showNumber()

#will return error for the user defined class object does not have logic to use operators "+,-,*,/" 

addNUM = num1 + num2  
addNUM.showNumber()

subNUM = num3 - num2
subNUM.showNumber()


    
````
