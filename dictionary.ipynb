## Working with single level dictionary

```python
employee = {'name':'talha', 'age':'29' , 'lang':'python'}
```

```python
employee['lang']= [employee['lang'],'js','java']
print(employee['lang'])
employee['lang'].append('c++')
print(employee['lang'],"only add 1value at the end ")
new_lang=['C#','ruby']
employee['lang'].extend(new_lang)
print(employee['lang'],'will add multi value')
```

```python

'''
Question:
You have the following dictionary in Python:

How can you add a new key-value pair "experience": 5 to this dictionary?
Write code to update the "salary" to 75000.
How can you check if the key "skills" exists in this dictionary?
What will employee.get("location", "Not Specified") return?
'''

employee = {
    "name": "John",
    "age": 30,
    "department": "IT",
    "skills": ["Python", "JavaScript", "SQL"],
    "salary": 70000
}
```

```python
print('How can you add a new key-value pair "experience": 5 to this dictionary?')
print(employee)
# dictionay_name['key'] = 'Value'
#employee['experience']='5'
dict_exp={'experience': '5'}
employee.update(dict_exp) # we use [] for adding new key
print(employee)
```

#### .update : Update the exisisting dictionary with the new pairs or with different dictionary

```python
print('Write code to update the "salary" to 75000.')
print(employee)
new_data= {
    'salary':'75000' # defining the new dictionary
}
employee.update(new_data) # updating the exisisting dictionary with new dictionary <for exisiting dict its updation > <for new dict is addition to the exisisting data>
'''employee['salary']='75000' '''
print(employee)
```

```python
print('How can you check if the key "skills" exists in this dictionary?')
print(employee)
# get methode > .get(key , )
if employee.get('skills', None):
    print("it does exist")
else :
    print("skills doesn't exist")
```

```python
print('What will employee.get("location", "Not Specified") return?')
print(employee.get("location", "Not Specified") )
```

```python

'''
Question:

You have the following nested dictionary representing data about a company:

Using this dictionary, answer the following:

Write code to fetch the names of all employees who are currently working on projects with the status "In Progress".
How can you retrieve all the unique skills of employees across the company (combine and deduplicate the skills from all employees)?
Find the name and role of the employee with the most projects assigned.
If a new project {"name": "Project E", "status": "Not Started"} is assigned to "Alice," update her projects list accordingly.
What code would you use to print the company name, location, and the total number of employees?
'''


company_data = {
    "company": "TechCorp",
    "employees": [
        {
            "id": 1,
            "name": "Alice",
            "role": "Developer",
            "skills": ["Python", "Django", "REST APIs"],
            "projects": [
                {"name": "Project A", "status": "Completed"},
                {"name": "Project B", "status": "In Progress"}
            ]
        },
        {
            "id": 2,
            "name": "Bob",
            "role": "Designer",
            "skills": ["Figma", "Sketch", "Photoshop"],
            "projects": [
                {"name": "Project C", "status": "Completed"},
                {"name": "Project D", "status": "Not Started"}
            ]
        },
        {
            "id": 3,
            "name": "Charlie",
            "role": "Manager",
            "skills": ["Leadership", "Communication", "Strategy"],
            "projects": []
        }
    ],
    "location": "New York",
    "established": 2010
}

```

```python
print(company_data.keys())
```

Normal ways

```python

# Write code to fetch the names of all employees who are currently working on projects with the status "In Progress".
all_emp = company_data.get('employees')
# print(type(company_data))
# print(all_emp)
# print(type(company_data.get('employees')))
# print(len(all_emp))

for employee in all_emp:
    # print(employee)
    # print(employee.get('name' ,'Not Found'))
    # print(employee.get('projects'))
    projects = employee.get('projects')
    for project in projects:
        #print(project['status'])
        if project['status']=='In Progress':
            print(employee.get('name'),project['status'])

    print()
```
