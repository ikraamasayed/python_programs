# List

```python
nums = [1,3,5,7,9]
# mutable
name =['Ikrama','Taha','Talha']
mix= [nums,name]#can have different type of elements
mix
```

##### Append
add the element at the end of the list

```python
print(nums)
nums.append(2)
nums
```

##### Insert
very similar to append peticularly at the specific index,
insert 4 at 2nd index
insert 6 at 4th index

```python
print(nums)
nums.insert(2,4) #insert 4 at the 2nd index
nums.insert(4,6) #insert 6 at 4th index
print(nums)
```

##### Remove  
 will search for 2 & remove @ from the list

```python
print(nums)
nums.remove(2)
nums
```

##### Pop
will Pop the element from that index as it was in the stack

```python
print(nums)
x=nums.pop(2)
print(x)
```

##### del
del multiple value

```python
print(nums)
del nums[-2:]
nums
```

##### Sort


```python
nums.insert(2,9)
print(nums)
nums.sort()
nums
```

##### Reverse Sort

```python
nums.reverse()
nums
```

Slicing list [start:stop:step]

```python
a = nums[1:5]
print(a)
```

```python
my_fruit = ['apple','banana','cherry','orange','kiwi','melon','mango']
ur_fruit = my_fruit
ur_fruit.append('pineapple')

if my_fruit == ur_fruit : print(f"same")

theirfruit = ur_fruit.copy()
theirfruit.append('strawberry')

if ur_fruit == theirfruit or my_fruit==theirfruit : print(f"same")
else: print(f"different using copy()")

these_fruit= my_fruit[:]
if these_fruit == my_fruit:  print(f"using slicing way to copy ")
```
