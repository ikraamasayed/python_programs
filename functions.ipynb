##### WAF for calculating the factorial of n! number

````python
def cal_fact (n):
    fact = 1    
    for i in range(1 , n+1):
        fact *=i
    return fact
print(cal_fact(5))
````

### Recursion 

````python
def print_number(n):
    if n==0:    # Base case : stops when reach 0 
        return(print("this is end"))    #if matched ends here
    
    print(n)    #print the current number
    print_number(n-1)   #Recursive call with a smaller value

print_number(10) # calling the function very first time
````

##### WARF to calculate the sum of first n natural numbers

````python
def calculate_sum(n):
    if n==0:
        return 0
    return calculate_sum(n-1)+n

calculate_sum(5)
````

### Anonymous function AKA Lamda functions

**What if ?** we just want to use it for once & even save the waste the lines of code 
#### in that scenario we use **Lambda function**

#### Filter = will filter from the iterable using specific function  
1. even is list
2. `list()` class has `filter()`  function 
3. in `filter()` filter function requires the 2 arg (function , none 0r Iterable   'lst,tup,dic,ect'   ) 

````python
nums= [1,2,3,4,5,6,7,8,9]

even= list(filter(lambda n:n%2==0,nums))
odd= list(filter(lambda n:n%2!=0,nums))
print(even)
print(odd)
````

#### Map = will Map & Update the iterable into the new variable
 

````python
nums= [1,2,3,4,5,6,7,8,9]

even= list(filter(lambda n:n%2==0,nums))
odd= list(filter(lambda n:n%2!=0,nums))

square = list(map(lambda n: n*n, nums))

print(nums)
print(square)
````

#### Reduce 
- reduce belonngs to the module functools 
- reduce takes 2 arg **( function , iterables )**
- reduce doesn't require the `list()` to contain 
- used to perfrom operations to get single value out of iterables

````python
from functools import reduce

nums= [1,2,3,4,5,6,7,8,9]

sum = reduce(lambda a,b : a+b,nums)
print(sum)
````
