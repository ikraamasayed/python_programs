- `def smart_div(func)` is a decorator for div

````python
def div (a,b):
    print (a/b)

div(2,8)

# what if the numerator < dinominator 
# make a logic : whatever will be the sequence numerator should be > denominator 

def smart_div(func):        #smart_div will be decorator for div & take the fucnction as parameter 

    def inner(x,y):         
        if x<y:
            x,y=y,x

            return func(x,y)
        
    return inner

div1= smart_div(div)
div1(2,8)
````

# 59. Operator Overloading

````python
class Students :

    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
    
    def showNumber (self):
        print(self.m1, self.m2)

    def __add__ (self,other):
        a1 = self.m1 + other.m1
        a2 = self.m2 + other.m2
        s3 = Students(a1,a2)

        return s3
    
    def __gt__ (self, other):       # use for greater than 
        n1 = self.m1 + self.m2
        n2 = other.m1 + other.m2

        if n1 > n2 :
            return True
        else :    
            return False 

S1 = Students(0,1)
S1.showNumber()
S2 = Students(2,3)
S2.showNumber()

S3 = S1 + S2
S3.showNumber()

if S1 > S2:
    print( "S1 is Winner")
else :
    print("S2 is Winner")
````

# 60. Method Overloading & Method Overriding

## Method Overloading

````python
class Students :
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2

    def sum(self,a = None, b = None , c = None):
        if a!= None and b!= None and c!= None:
            s = a+b+c
        elif a!= None and b!= None:
            s = a+b
        else:
            s = a
        return s 

S1 = Students(8,5)
print(S1.sum(5,3))
````

# **Notes on [MultiThreading](https://youtu.be/GqHLztqy0PU?si=Xoo5eeaLqbGZ9n0M) in Python**  

## **1. Introduction to Threading**  
Threading is a technique that allows a program to run multiple operations concurrently in the same process. Each independent sequence of execution is called a **thread**. Using threading, we can improve the efficiency of programs, especially for tasks that involve waiting (e.g., network requests, file operations, or user interactions).  

---

## **2. MultiThreading**  
**MultiThreading** is the process of running multiple threads simultaneously. It is useful when a program needs to perform multiple tasks at the same time, such as running background tasks without blocking the main execution.  

In Python, the **threading** module provides a way to create and manage threads.  

---

## **3. Understanding the Code: Before and After MultiThreading**  

### **Before MultiThreading (Sequential Execution)**  

````python
from time import sleep #using time to delay 
class Hello :
    def run (self):
        for i in range (5):
            print("Hello")
    
class Hi:
    def run(self):
        for i in range (5):
            print("Hi")
        
H1= Hello()
H2 = Hi()

H1.run()    #execute first 
H2.run()    #execute after H1 completes
````

**Problems in the "Before" Code**      **VS**      **"After" Code**
- The run() method executes the functions sequentially.     **VS**    **.** Hello and Hi execute simultaneously instead of one after another.

- The Hello class runs first, and only after it finishes, the Hi class starts.      **VS**    **.** sleep(0.1) is used to simulate delays in execution, demonstrating the parallel nature of threads.

- This approach does not take advantage of concurrent execution.        **VS**    **.** We use start() instead of run() to create actual threads.

````python
from threading import Thread
from time import sleep

class Hello(Thread):
    def run(self):
        for i in range(10):
            print("Hello")
            sleep(0.2)  # Adding a delay to simulate a real scenario

class Hi(Thread):
    def run(self):
        for i in range(10):
            print("Hi")
            sleep(0.1)

H1 = Hello()
H2 = Hi()

H1.start()  # Starts a new thread for H1
#sleep(0.2)  # Small delay to allow Hello to start first
H2.start()  # Starts a new thread for H2
````

# **[Binary Search](https://youtu.be/DE-ye0t0oxE?si=gnu4X8e0wqcB-JZO)**

````python
position = 0
def search (arr,n):
    l = 0
    u = len(arr)-1

    while l <= u:
        mid = (l+n)//2
    
        if arr[mid]==n:
            globals()['position'] = mid
            return True
        else:
            if arr[mid]<n:
                l = mid+1
            else :
                u = mid-1
    return False

arr = [1,3,5,7,9,2,4,6,8,0]
arr.sort()
n = 8 

if search(arr,n):
    print("Found ",position+1)
else:
    print("not found")
````

# **[Bubble Sort](https://youtu.be/Vca808JTbI8?si=cveFJx9LpuM9zyal)**

````python
def bbsort(nums):
    for i in range (len(nums)-1,0,-1):
        for j in range (i):
            if nums[j]>nums[j+1]:
                temp= nums[j]
                nums[j] = nums[j+1]
                nums[j+1] = temp

list_of_nums = [2,4,6,8,0,1,3,5,7,9]
print(list_of_nums)
bbsort(list_of_nums)

print(list_of_nums)
````
