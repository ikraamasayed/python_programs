find the x number and it's index position in the list or tuple

```python
tu= (1,4,9,36,16,25,36,49,64,81,100,36) #given list
x=36 # value to find 
att= 0 # initialization 
while att <len(tu): # conditon 
    if tu[att]==x:  # if tu[3]/tu[6]/tu[11] '36' == x '36'
        print(x,"found at",att," index")
    else:
        print("finding....")
    att+=1 # incrementation 
```

### Break & Continue

Break: used to terminate the loop when encountered

```python
tu= (1,4,9,36,16,25,36,49,64,81,100,36) #given list
x=36 # value to find 
att= 0 # initialization 
while att <len(tu): # conditon 
    if tu[att]==x: 
        print(x,"found at",att," index")
        break
    else:
        print("finding....")
    att+=1 # incrementation 
print("Found once requierement done ")
```

Continue: terminates execution in the current iteration & continues execution of the loop with the next iteration.

```python
i = 0 
while i <=7:
    if (i ==3): # here if 1=3 increment it by 1 
        i+=1
        continue #skips the futher iteration which is prin(i)
    print(i)
    i+=1
    
```

#### Range ()

Range functions returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.

range(start ? , stop, step ? )          (starts at 0 *optinal*  ,end at 20 , skips 4 steps *optional* )

```python
for el_1 in range (8):
    print (el_1)
print( "1 value is only for the stop value")
print()


print("start from 24")
for el in range(24,30):
    print(el )
print("stop value is 30")
print()


print("start from 60")
for el in range (60,80,4):
    print(el)
print("stop at 80, skip or take step of 4")
    
```
