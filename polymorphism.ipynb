# Operator Overloading

````python
class Students :

    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
    
    def showNumber (self):
        print(self.m1, self.m2)

    def __add__ (self,other):
        a1 = self.m1 + other.m1
        a2 = self.m2 + other.m2
        s3 = Students(a1,a2)

        return s3
    
    def __gt__ (self, other):       # use for greater than 
        n1 = self.m1 + self.m2
        n2 = other.m1 + other.m2

        if n1 > n2 :
            return True
        else :    
            return False 
        

        


S1 = Students(0,1)
S1.showNumber()
S2 = Students(2,3)
S2.showNumber()

S3 = S1 + S2
S3.showNumber()

if S1 > S2:
    print( "S1 is Winner")
else :
    print("S2 is Winner")
````

# Method Overloading

````python
class Students :
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2

    def sum(self,a = None, b = None , c = None):
        if a!= None and b!= None and c!= None:
            s = a+b+c
        elif a!= None and b!= None:
            s = a+b
        else:
            s = a
        return s 

S1 = Students(8,5)
print(S1.sum(5,3))
````

# Method Overriding 

`Son` actually don't have any phone attribute but inherites the `Dad's` phone's Attribute since he don't have any 
Whereas `Elder-son` has `iPhone` even tho he inherite's `Dad` will show his own `iPhone`

overriding the inherited value as soon as it get it's own value

````python
class Dad:
    def phone(self):
        print("Have Motorola")

class Elder_Son(Dad):
    def phone(self):
        print("Have iPhone")
class Son(Dad):
    pass
S1 = Elder_Son()
S2= Son()
S1.phone()
S2.phone()
````
