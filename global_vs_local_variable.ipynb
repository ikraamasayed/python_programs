## Global Variable 
are those variable which are generally outside the scope 

```python
a = 10 
print(a,"id =",id(a))
def something():
    print(a,"id =",id(a),"this is global `a` inside the local function")

something()
```

## Local Variable 
are varianbles which are stored inside of the scope of perticular function , method ,etc

```python
a = 10 
print(a,", id =",id(a))
def something():
    a = 50
    print(a,", id =",id(a),"this is local `a`")

something()
```

## Global Inside the Local variable

- Adding `global a` inside the scope will set the local value to the global variable

```python
a = 10 
print(a,", id =",id(a),'\ncalled at global scope,\n')
def something():
    global a
    a = 50

    print(a,", id =",{id(a)},"this is local `a` called inside func\nat the local scope\n")

something()

print(a,", id =",id(a),"this is globla `a` called outside func")
print('local {a} i.e 50 now = global {a}')
```

- setting `globals ()['a'] = 37` will chng the global value from inside the scope of local 

```python
a = 10 

print(a,", id =",id(a),'        called at global scope,\n')
def something():
    a = 50
    
    globals()['a']=37

    print(a,", id =",id(a),"this is local value \n")
    
something()

print(a,", id =",id(a),"this global var was chng from inside the local variable `a` called outside func \n")
print('global "a" i.e 10 now = global',a)
```
